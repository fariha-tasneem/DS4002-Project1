```python
import pandas as pd

# Read the 2 CSV files
df1 = pd.read_csv('/Users/farihatasneem/Downloads/Books_rating.csv')
df2 = pd.read_csv('/Users/farihatasneem/Downloads/books_data.csv')
```

```python
# MERGE THE 2 CSV FILES 
columns_to_combine = ['Title', 'description', 'authors', 'publisher', 'publishedDate', 'categories', 'ratingsCount'] 
common_column = 'Title'
combined_df = df1.merge(df2[columns_to_combine], on=common_column, how='inner')
combined_df
```

```python
# DROP ROWS WITH MISSING DATA
combined_df.dropna(inplace=True)
combined_df
```

```python
# Group by 'categories' and calculate the average ratings for each genre
genre_ratings = combined_df.groupby('categories')['review/score'].mean().reset_index()

# Sort the genres by average ratings in ascending order to find the lowest ratings
lowest_rated_genres = genre_ratings.sort_values('review/score')

# Display the genres with the lowest average ratings
print(lowest_rated_genres)
```

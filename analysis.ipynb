```python
import pandas as pd

# Read the CSV file
df_top_30_genres = pd.read_csv('top_30_genres.csv')
```

```python
df_top_30_genres
```

```python
top_30_genres = df_top_30_genres['categories'].value_counts().head(30).index.tolist()
print(top_30_genres)
```

```python
# HOW MANY REVIEWS ARE IN THE 30TH GENRE?

thirtieth_genre = top_30_genres[29]

genre_30_df = df_top_30_genres[df_top_30_genres['categories'] == thirtieth_genre]

num_reviews_in_30th_genre = len(genre_30_df)

print(num_reviews_in_30th_genre)
```

```python
# TO RANDOM SAMPLE 1100 REVIEWS FROM THE TOP 30 GENRES 

import pandas as pd

# initialize an empty DataFrame to store the random samples
random_samples_df = pd.DataFrame()

num_reviews_to_sample = 1100

# loop through each genre and sample 1100 reviews
for genre in top_30_genres:
    # Filter the DataFrame for the current genre
    genre_df = df_top_30_genres[df_top_30_genres['categories'] == genre]
    
    if len(genre_df) >= num_reviews_to_sample:
        # Sample reviews randomly
        random_sample = genre_df.sample(n=num_reviews_to_sample, random_state=42) 
        
        # Append the random sample to the results DataFrame
        random_samples_df = pd.concat([random_samples_df, random_sample], ignore_index=True)
    else:
        random_samples_df = pd.concat([random_samples_df, genre_df], ignore_index=True)

random_samples_df = random_samples_df.reset_index(drop=True)

random_samples_df['categories'] = random_samples_df['categories'].str.strip('[]')
```

```python
random_samples_df
```

```python
# VISUALIZE HOW MANY BOOKS BELONG IN EACH GENRE/CATEGORY

import matplotlib.pyplot as plt
import seaborn as sns

genre_counts = random_samples_df['categories'].value_counts()
# bar chart
plt.figure(figsize=(10, 6))  

sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.xticks(rotation=90) 
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Distribution of Book Genres')
plt.tight_layout()

plt.show()
```

```python
# CALCUALTE AVERAGE RATINGS BY GENRE AND DISPLAY PLOT FOR AVERAGE RATINGS

genre_ratings = random_samples_df.groupby('categories')['review/score'].mean()
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  
sns.barplot(x=genre_ratings.index, y=genre_ratings.values, palette='viridis')
plt.xticks(rotation=90)  
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Ratings by Genre')
plt.tight_layout()
plt.show()
```

```python
# CREATE COLUMN FOR SENTIMENT ANALYSIS SCORE
from textblob import TextBlob
import pandas as pd

def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

random_samples_df['sentiment_score'] = random_samples_df['review/text'].apply(get_sentiment)

```

```python
# INITIAL SENTIMENT ANALYSIS SCORE ACROSS GENRES
# BOXPLOT OF SENTIMENT ANALYSIS SCORES

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

sns.boxplot(x='categories', y='sentiment_score', data=random_samples_df, palette='Set3')

plt.xticks(rotation=90)  
plt.xlabel('Genre')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Scores of Reviews Across Genres')

plt.show()
```

```python
random_samples_df['year'] = random_samples_df['publishedDate'].str.extract(r'(\d{4})')[0]
random_samples_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fiction_df = random_samples_df[random_samples_df['categories'] == "'Fiction'"]

yearly_sentiment_mean = fiction_df.groupby('year')['sentiment_score'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(yearly_sentiment_mean['year'], yearly_sentiment_mean['sentiment_score'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Mean Sentiment Score')
plt.title('Mean Sentiment Score Trend for Fiction Genre Over the Years')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Read the 2 CSV files
df1 = pd.read_csv('/Users/farihatasneem/Downloads/Books_rating.csv')
df2 = pd.read_csv('/Users/farihatasneem/Downloads/books_data.csv')
```

```python
# MERGE THE 2 CSV FILES 
columns_to_combine = ['Title', 'description', 'authors', 'publisher', 'publishedDate', 'categories', 'ratingsCount'] 
common_column = 'Title'
combined_df = df1.merge(df2[columns_to_combine], on=common_column, how='inner')
combined_df
```

```python
# DROP ROWS WITH MISSING DATA
combined_df.dropna(inplace=True)
combined_df
```

```python
# SUMMARY STATISTICS
summary_stats = combined_df.describe()
summary_stats
```

```python
# ONLY GET THE TOP 10 GENRES IN THE DATASET
top_10_genres = combined_df['categories'].value_counts().head(10).index.tolist()
df_top_10_genres = combined_df[combined_df['categories'].isin(top_10_genres)]
df_top_10_genres
```

```python
# ONLY GET THE TOP 20 GENRES IN THE DATASET
top_20_genres = combined_df['categories'].value_counts().head(20).index.tolist()
df_top_20_genres = combined_df[combined_df['categories'].isin(top_20_genres)]
df_top_20_genres
```

```python
# RANDOM SAMPLING 10 GENRES

import matplotlib.pyplot as plt
import seaborn as sns
import random  


random_genres = random.sample(list(combined_df['categories']), k=10)  

# Dataframe for random genres
df_random_genres = combined_df[combined_df['categories'].isin(random_genres)]

# number of books in each category
genre_counts = df_random_genres['categories'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Distribution of Random Book Genres')
plt.tight_layout()

plt.show()
```

```python
# ONLY GET THE TOP 30 GENRES IN THE DATASET
top_30_genres = combined_df['categories'].value_counts().head(30).index.tolist()
df_top_30_genres = combined_df[combined_df['categories'].isin(top_30_genres)]
df_top_30_genres
```

```python

file_path = 'top_30_genres.csv'

df_top_30_genres.to_csv(file_path, index=False)
```

```python
# HOW MANY REVIEWS ARE IN THE 30TH GENRE?

thirtieth_genre = top_30_genres[29]

genre_30_df = df_top_30_genres[df_top_30_genres['categories'] == thirtieth_genre]

num_reviews_in_30th_genre = len(genre_30_df)

print(num_reviews_in_30th_genre)
```

```python
# TO RANDOM SAMPLE 1100 REVIEWS FROM THE TOP 30 GENRES 

import pandas as pd

# initialize an empty DataFrame to store the random samples
random_samples_df = pd.DataFrame()

num_reviews_to_sample = 1100

# loop through each genre and sample 1100 reviews
for genre in top_30_genres:
    # Filter the DataFrame for the current genre
    genre_df = df_top_30_genres[df_top_30_genres['categories'] == genre]
    
    if len(genre_df) >= num_reviews_to_sample:
        # Sample reviews randomly
        random_sample = genre_df.sample(n=num_reviews_to_sample, random_state=42) 
        
        # Append the random sample to the results DataFrame
        random_samples_df = pd.concat([random_samples_df, random_sample], ignore_index=True)
    else:
        random_samples_df = pd.concat([random_samples_df, genre_df], ignore_index=True)

```

```python
# CREATE COLUMN FOR SENTIMENT ANALYSIS SCORE
from textblob import TextBlob
import pandas as pd

def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

random_samples_df['sentiment_score'] = random_samples_df['review/text'].apply(get_sentiment)

```

```python
# INITIAL SENTIMENT ANALYSIS SCORE ACROSS GENRES
# BOXPLOT OF SENTIMENT ANALYSIS SCORES

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

sns.boxplot(x='categories', y='sentiment_score', data=random_samples_df, palette='Set3')

plt.xticks(rotation=90)  
plt.xlabel('Genre')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Scores of Reviews Across Genres')

plt.show()

```

```python
# BARCHART OF SENTIMENT ANALYSIS SCORES

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

sns.barplot(x='categories', y='sentiment_score', data=random_samples_df, palette='Set3')

plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Average Sentiment Score')
plt.title('Average Sentiment Scores of Reviews Across Genres')

plt.show()
```

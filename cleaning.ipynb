```python
import pandas as pd

# Read the 2 CSV files
df1 = pd.read_csv('/Users/farihatasneem/Downloads/Books_rating.csv')
df2 = pd.read_csv('/Users/farihatasneem/Downloads/books_data.csv')
```

```python
# MERGE THE 2 CSV FILES 
columns_to_combine = ['Title', 'description', 'authors', 'publisher', 'publishedDate', 'categories', 'ratingsCount'] 
common_column = 'Title'
combined_df = df1.merge(df2[columns_to_combine], on=common_column, how='inner')
combined_df
```

```python
# DROP ROWS WITH MISSING DATA
combined_df.dropna(inplace=True)
combined_df
```

```python
# SUMMARY STATISTICS
summary_stats = combined_df.describe()
summary_stats
```

```python
# ONLY GET THE TOP 10 GENRES IN THE DATASET
top_10_genres = combined_df['categories'].value_counts().head(10).index.tolist()
df_top_10_genres = combined_df[combined_df['categories'].isin(top_10_genres)]
df_top_10_genres
```

```python
# VISUALIZE HOW MANY BOOKS BELONG IN EACH GENRE/CATEGORY

import matplotlib.pyplot as plt
import seaborn as sns

genre_counts = df_top_10_genres['categories'].value_counts()
# bar chart
plt.figure(figsize=(10, 6))  

sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.xticks(rotation=90) 
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Distribution of Book Genres')
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import random  


random_genres = random.sample(list(combined_df['categories']), k=10)  

# Dataframe for random genres
df_random_genres = combined_df[combined_df['categories'].isin(random_genres)]

# number of books in each category
genre_counts = df_random_genres['categories'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Distribution of Random Book Genres')
plt.tight_layout()

plt.show()
```

```python
# CREATE COLUMN FOR SENTIMENT ANALYSIS SCORE
from textblob import TextBlob
import pandas as pd

def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

df_top_10_genres['sentiment_score'] = df_top_10_genres['review/text'].apply(get_sentiment)

```

```python
# INITIAL SENTIMENT ANALYSIS SCORE ACROSS GENRES

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

sns.boxplot(x='categories', y='sentiment_score', data=df_top_10_genres, palette='Set3')

plt.xticks(rotation=90)  
plt.xlabel('Genre')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Scores of Reviews Across Genres')

plt.show()

```
